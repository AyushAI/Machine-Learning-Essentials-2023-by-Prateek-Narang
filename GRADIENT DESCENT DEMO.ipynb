```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

GENERATE THE DATA USING A DUMMY FUNCTION

```python
X = np.arange(10)
print(X)
```

```python
Y = (X-5)**2+3
```

```python
Y
```

VISUALIZE THE DATA

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

GRADIENT DESCENT

```python
# initialize the x with any random variable
x = 0
alpha = 0.1
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x - alpha * grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    
plt.show()
```
